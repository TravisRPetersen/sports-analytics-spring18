## SQL Select Tutorial

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sqlite3
from scipy import stats
plt.style.use('ggplot')
```

```python
con = sqlite3.connect("lahmansbaseballdb.sqlite")
```

```python
pd.read_sql('select * from teams LIMIT 5', con=con).head()
```

```python
pd.read_sql('select teamID, yearID, H, HR, R, AB from teams where yearID > 1999', con=con).head()
```

```python
pd.read_sql('select teamID, yearID, H, HR, R, AB from teams where yearID > 1999 and lgID = "AL"', con=con).head()
```

```python
pd.read_sql('''SELECT teamID, 
                      yearID, 
                      H, 
                      HR, 
                      R, 
                      AB 
               FROM teams 
               WHERE yearID > 1999 
                   AND lgID = "AL" 
                   AND (teamID = "BAL" OR teamID="NYA") 
''', con=con).head()
```

```python
pd.read_sql('''
    SELECT teamID, 
           yearID, 
           H, 
           HR, 
           R, 
           AB 
    FROM teams 
    WHERE yearID > 1999 
        AND teamID in ("BAL", "NYA", "BOS", "TOR", "TB")
''', con=con).head()
```

```python
pd.read_sql(''' SELECT teamID, 
                       yearID, 
                       H, 
                       HR, 
                       R, 
                       AB 
                FROM teams 
                WHERE yearID>1999 
                    AND teamID like "TB%";
''', con=con).head()
```

```python
pd.read_sql('''
        SELECT teamID, 
               yearID, 
               H, 
               HR, 
               R, 
               AB
        FROM teams 
        WHERE yearID > 1999 
            AND teamID in ("BAL", "NYA", "BOS", "TOR", "TBA")
    
''', con=con).head()
```

```python
pd.read_sql(
'''SELECT teamID, 
          yearID, 
          H, 
          HR, 
          R, 
          AB 
    FROM teams 
    WHERE yearID > 1999 and teamID in ("BAL", "NYA", "BOS", "TOR", "TBA")
    ORDER by HR 
''', con=con).head()
```

```python
pd.read_sql(
''' 
    SELECT teamID, 
           yearID, 
           H, 
           HR,
           R, 
           AB 
    FROM teams 
    WHERE yearID > 1999 
        AND teamID in ("BAL", "NYA", "BOS", "TOR", "TBA")
    ORDER by HR DESC
''', con=con).head()
```

```python
pd.read_sql(
''' SELECT teamID, 
           H, 
           sum(HR), 
           count(HR) 
    FROM teams 
    WHERE yearID > 1995 
        AND teamID in ("BAL", "NYA", "BOS", "TOR", "TBA")
    GROUP BY teamID 
    ORDER by sum(HR) DESC
''', con=con).head()
```

```python
pd.read_sql('''
    SELECT teamID, 
            sum(HR), 
            count(HR), 
            sum(HR)/count(HR) as avg_HR from teams 
    WHERE yearID > 1999 
        AND teamID in ("BAL", "NYA", "BOS", "TOR", "TBA")
    GROUP BY teamID 
    ORDER BY sum(HR) DESC
''', con=con).head()
```

```python
pd.read_sql('''
    SELECT teamID, 
           sum(HR) as sum_HR, 
           count(HR) as seasons, 
           sum(HR)/count(HR) as avg_HR from teams 
    WHERE yearID > 1999 
        AND teamID in ("BAL", "NYA", "BOS", "TOR", "TBA")
    GROUP BY teamID 
    HAVING sum(HR)>2800
''', con=con).head()
```
