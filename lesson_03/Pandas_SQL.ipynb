```python
import pandas as pd
import numpy as np
```

```python
top_16_draft = pd.read_csv('top_16_draft.csv', usecols=['Rk','Pk','Tm', 'Player', 'College'])
```

```python
nba_stats = pd.read_table('nba_16.txt', sep=',')
```

```python
top_16_draft.head()
```

```python
nba_stats.head()
```

## Outer Join

```python
outer_merge = pd.merge(top_16_draft, nba_stats, left_on='Player', right_on='Name', how='outer')
outer_merge.head()
```

```python
outer_merge.tail()
```

## Right Join

```python
right_merge = pd.merge(top_16_draft, nba_stats, left_on='Player', right_on='Name', how='right')
right_merge.head()
```

```python
right_merge.tail()
```

## Left Join

```python
left_merge = pd.merge(top_16_draft, nba_stats, left_on='Player', right_on='Name', how='left')
left_merge.head()
```

```python
left_merge.tail()
```

## Inner Join

```python
inner_merge = pd.merge(top_16_draft, nba_stats, left_on='Player', right_on='Name', how='inner')
inner_merge
```

## Pandas Concat

```python
nba_stats_missing = pd.read_table('nba_16_pt2.txt', sep=',')
```

```python
nba_all = pd.concat([nba_stats, nba_stats_missing])
```

## Final Left Merge

```python
left_merge = pd.merge(top_16_draft, nba_all, left_on='Player', right_on='Name', how='left')
left_merge
```

## Pandas Split Apply Combine

Sources:

* **Splitting** the data into groups based on some criteria

* **Applying** a function to each group independently

* **Combining** the results into a data structure

```python
by_team = nba_all.groupby('Tm')
by_team
```

Calling count returns the total number of NOT NULL values within each column. If we were interested in the total number of records in each group, we could use size.

```python
print(by_team.count().head()) # NOT NULL records within each column
print('\n')
print(by_team.size().tail()) # total records for each department
```

Summation can be done via sum, averaging by mean, etc. (if it's a SQL function, chances are it exists in pandas). Oh, and there's median too, something not available in most databases.

```python
print(by_team.sum()[20:25]) # total salaries of each department
print('\n')
print(by_team.mean()[20:25]) # average salary of each department
print('\n')
print(by_team.median()[20:25]) # take that, RDBMS!
```

Operations can also be done on an individual Series within a grouped object. Say we were curious about the five teams with the most distinct ages - the pandas equivalent to:

```python
by_team.Age.nunique().sort_values(ascending=False)[:5]
```
