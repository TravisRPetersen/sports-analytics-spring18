## Boolean Slicing

Rules for specifying multiple filter criteria in pandas:

* use & instead of and
* use | instead of or
* add parentheses around each condition to specify evaluation order

```python
hr_df[hr_df['Name']!='Sammy Sosa'].head()
```

```python
hr_df[(hr_df.Name != 'Sammy Sosa') & (hr_df.Season == 2001)]
```

Using a tilda (~) gives the ability to reverse the condition

```python
hr_df[~(hr_df.Name=='Sammy Sosa')].head()
```

## Lambda Function
* If we try to take the mean of a field stored as a string we will get an error
* We can use a lambda function to remove the % from the number and then convert to a float

```python
hr_df['BB%'].mean()
```

```python
hr_df['BB_PCT'] = hr_df['BB%'].apply(lambda x: float(x[:-1]))
```

```python
hr_df['BB_PCT'].mean()
```
